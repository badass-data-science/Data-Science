# ETL: Load Oanda Candlestick Data Into InfluxDB

This is intended to be run using papermill (https://papermill.readthedocs.io/en/latest/).

## Load useful libraries

```python
import os
```

```python
from python_tools_and_shortcuts.databases.influxdb.InfluxDbTool import InfluxDbTool

from forex.etl.pipelines.CandlestickPipeline import CandlestickPipeline
from forex.etl.config.schema_config import ALLOWED_TAGS, ALLOWED_FIELDS
from forex.etl.config.database_config import INFLUXDB_URL, INFLUXDB_TOKEN, INFLUXDB_ORG
from forex.etl.config.database_config import INFLUXDB_BUCKET
```

## User settings

```python
run_test_query = False
```

Intended to be set by papermill:

```python
config_file = os.environ['CREDS']
instrument = 'EUR/USD'
granularity = 'H1'
```

## Open a connection to InfluxDB

```python
ifc = InfluxDbTool(INFLUXDB_URL, INFLUXDB_TOKEN, INFLUXDB_ORG)
```

## Instantiate the pipeline

```python
cp = CandlestickPipeline(
    config_file,
    instrument,
    granularity,
    ifc,
    ALLOWED_TAGS = ALLOWED_TAGS,
    ALLOWED_FIELDS = ALLOWED_FIELDS,
    INFLUXDB_BUCKET = INFLUXDB_BUCKET,
    run_test_query = run_test_query,
)
```

## Run the pipeline

```python
cp.fit()
```

## Close the database connection

```python
del(ifc)
```
