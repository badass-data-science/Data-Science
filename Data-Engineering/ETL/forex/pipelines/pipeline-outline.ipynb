# Basic Pipeline

## Load useful libraries

```python
import os
import papermill as pm
```

## User settings (papermill)

```python
instrument = 'EUR/USD'
granularity = 'H1'
config_file = os.environ['CREDS']
n_back = 200
lookahead = 4
```

## User settings (intentionally hard-coded)

```python
granularity_list_ff = ['H1', 'M15']
```

## Retrieve data from Oanda's API

```python
result = pm.execute_notebook(
    os.environ['FOREX_HOME'] + '/forex/etl/pipelines/papermill/ETL-Oanda-candlestick-data.ipynb',
    '/tmp/ETL-Oanda-candlestick-data_' + instrument.replace('/', '_') + '_' + granularity + '.ipynb',
    parameters = 
        {
            'instrument': instrument,
            'granularity': granularity,
            'config_file' : config_file,
        },
)
```

## Forward Fill

```python
if granularity in granularity_list_ff:
    result = pm.execute_notebook(
        os.environ['FOREX_HOME'] + '/forex/etl/pipelines/papermill/prepare-ff-data.ipynb',
        '/tmp/prepare-ff-data_' + instrument.replace('/', '_') + '_' + granularity + '.ipynb',
        parameters = 
            {
                'instrument': instrument,
                'granularity': granularity,
            },
    )
```
